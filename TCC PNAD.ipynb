```python
import pandas as pd, statsmodels.api as sm
```

```python
dados = pd.read_csv('dados.csv')

base = dados.copy()
dados
```

```python
# Filtrando apenas as colunas com os dados relevantes pra minha análise

colunas_drop = ['id_regiao',
 'id_uf', 'numero_familia',
 'dia_nascimento',
 'mes_nascimento',
 'ano_nascimento','sabe_ler_escrever',
 'frequenta_escola',
 'serie_frequentada',
 'grau_frequentado','trabalhou_semana',
 'tinha_trabalhado_semana',
 'tinha_outro_trabalho',
 'ocupacao_semana',
 'atividade_ramo_negocio_semana',
 'atividade_ramo_negocio_anterior',
 'possui_carteira_assinada',
 'renda_mensal_dinheiro',
 'renda_mensal_produto_mercadoria',
 'horas_trabalhadas_semana',
 'renda_mensal_dinheiro_outra',
 'renda_mensal_produto_outra',
 'horas_trabalhadas_outros_trabalhos',
 'contribui_previdencia',
 'tipo_instituto_previdencia',
 'tomou_providencia_conseguir_trabalho_semana',
 'tomou_providencia_ultimos_2_meses',
 'qual_providencia_tomou',
 'tinha_carteira_assinada_ultimo_emprego',
 'renda_aposentadoria',
 'renda_pensao',
 'renda_abono_permanente',
 'renda_aluguel',
 'renda_outras',
 'renda_mensal_ocupacao_principal',
 'renda_mensal_todos_trabalhos',
 'renda_mensal_todas_fontes',
 'atividade_ramo_negocio_agregado',
 'horas_trabalhadas_todos_trabalhos','renda_mensal_familia',
 'ocupacao_ano_anterior',
 'renda_mensal_dinheiro_deflacionado',
 'renda_mensal_produto_mercadoria_deflacionado',
 'renda_mensal_dinheiro_outra_deflacionado',
 'renda_mensal_produto_mercadoria_outra_deflacionado',
 'renda_mensal_ocupacao_principal_deflacionado',
 'renda_mensal_todos_trabalhos_deflacionado',
 'renda_mensal_todas_fontes_deflacionado',
 'renda_mensal_familia_deflacionado',
 'renda_aposentadoria_deflacionado',
 'renda_pensao_deflacionado',
 'renda_abono_deflacionado',
 'renda_aluguel_deflacionado',
 'renda_outras_deflacionado',
 'ultima_serie_frequentada', 'ultimo_grau_frequentado',
 'posicao_ocupacao', 'grupos_ocupacao', 'regiao_metropolitana', 'condicao_domicilio', 'numero_membros_familia', 'ordem']

dados2 = base.drop(columns=colunas_drop, axis=1)
dados2
```

```python
# Filtrando apenas os dados de Sergipe (região de análise)

dados = dados2.loc[dados2['sigla_uf'] == 'SE']
dados
```

```python
# Este filtro é encontrado no dicionário das PNADs. Condição de domicílio 1 seria pra responsável pela casa, 2 cônjuge e 3
# seria o filho. Embora o responsável pelo domicílio nem sempre seja o "pai" da família, utilizamos essa variável como proxy.

dados = dados.loc[dados['condicao_familia'] <= 3]
dados
```

```python
# Separando os anos de análise. 2010 e 2000 não possuíram PNADs por serem anos censitários, o que justifica a escolha do ano
# subsequente para manter a análise dentro da PNAD.

d2001 = dados.loc[dados['ano'] == 2001]
d2005 = dados.loc[dados['ano'] == 2005]
d2011 = dados.loc[dados['ano'] == 2011]
d2015 = dados.loc[dados['ano'] == 2015]

dados2 = pd.concat([d2001, d2005, d2011, d2015])
dados2.reset_index(drop=True, inplace=True)
dados2
```

```python
# Filtrando apenas as condições de domicílio para "pais e mães" (novamente, assumindo que responsáveis e conjuges sejam o casal)
dados_pais = dados2.loc[dados2['condicao_familia'] <= 2]
dados_pais
```

```python
# Tratamento descrito no TCC para utilizar a média dos anos de estudo dos pais como um dos betas da regressão.
dados_pais['med_anest_pais'] = dados_pais.groupby('id_domicilio')['anos_estudo'].transform('mean')
dados_pais
```

```python
# Criando um dicionário baseado no ID do domicílio para atribuir a média dos anos de estudo para pai e mãe
dict1 = dados_pais.groupby(['id_domicilio'])["anos_estudo"].mean()
dict1 = dict1.to_dict()
dict1
```

```python
# Criando coluna com média dos anos de estudo dos pais
dados2['anestpais_mean'] = dados2['id_domicilio'].map(dict1)
dados2
```

```python
# Criando um dict para transformar as dummies de raça em apenas brancos e não-brancos, evitando problemas de multicolinearidade.
dict_raca = {2:0, 4:1, 6:1, 8:1, 9:1}
dados2['raca_cor2'] = dados2['raca_cor'].map(dict_raca)
dados2
```

```python
# Variáveis de controle da idade dos filhos 
dados2['id_filhos'] = 15
dados2['id_filhos_quad'] = 225
dados2
```

```python
# Separando novamente os anos em diferentes dfs para realizar a regressão por ano 
d2001 = dados2.loc[dados2['ano'] == 2001]
d2005 = dados2.loc[dados2['ano'] == 2005]
d2011 = dados2.loc[dados2['ano'] == 2011]
d2015 = dados2.loc[dados2['ano'] == 2015]
```

```python
# Melhorando a interpretação do df renomeando a coluna
dados2.rename(columns = {'anestpais_mean':'med_anest_pais'}, inplace = True)
dados2
```

```python
# Modelo para o ano de 2001

X = d2001[['med_anest_pais', 'sexo', 'raca_cor2', 'idade_filho', 'idade_filho_quad']]
Y = d2001['anos_estudo']

modelo4 = sm.OLS(Y, X).fit()

modelo4.summary()
```

```python
# Modelo para o ano de 2005

X = d2005[['med_anest_pais', 'sexo', 'raca_cor2', 'idade_filho', 'idade_filho_quad']]
Y = d2005['anos_estudo']

modelo4 = sm.OLS(Y, X).fit()

modelo4.summary()
```

```python
# Modelo para o ano de 2011

X = d2011[['med_anest_pais', 'sexo', 'raca_cor2', 'idade_filho', 'idade_filho_quad']]
Y = d2011['anos_estudo']

modelo4 = sm.OLS(Y, X).fit()

modelo4.summary()
```

```python
# Modelo para o ano de 2015

X = d2015[['med_anest_pais', 'sexo', 'raca_cor2', 'idade_filho', 'idade_filho_quad']]
Y = d2015['anos_estudo']

modelo4 = sm.OLS(Y, X).fit()

modelo4.summary()
```
